```python
# Momentum
momLabels = momLabels_Dict[phaseInfo]
momValue  = momValue_Dict[phaseInfo]

def momInfo(mLab):
    mVal  = momValue[mLab]
    mTag = 'momXYZ.%d.%d.%d' % (momValue[mLab][0],momValue[mLab][1],momValue[mLab][2])
    return mVal, mTag
#------------------------------------------------
        
# Source and sink operators

# Read operators for given momentum
Nop_dict = {}
optrList = {}
for pL in momLabels:
    runTag = '%s_%s'%(phaseInfo,pL)
    operator_list = operatorList_Dict[runTag]
    optrList[runTag] = []
    with open(operator_list) as fp:
        line = fp.readlines()
        for t in line:
            optrList[runTag].append(t.split()[0])
    Nop_dict[runTag] = len(optrList[runTag])

print('Operator List:')
print(optrList)
#------------------------------------------------

# Directories, files
corr_dir='../2pt_corr_effEnergy/2pt_data_%s' % (phaseInfo)
#------------------------------------------------

# Ensemble parameters
L = 32
T = 64
binsize=1
#------------------------------------------------

# Nucleon, pion mass
alat_fm = 0.094
alat_iGeV = alat_fm / 0.1973
mpi_GeV = 0.350
mN_GeV = 1.123
amN = mN_GeV * alat_iGeV

# Dispersion relation
Pi = np.pi
def dispRel(am,k):
    ap = 2*Pi/L * k
    return np.sqrt(am**2 + np.dot(ap,ap))
    

print('\nDone\n')
```

## Jackknife functions

```python
def jackknife_ave(bins,Nbins,tL):

    Jax = 0
    
    if(tL==1):
        ave = np.mean(bins)

        sqsum = sum(map(lambda x:x*x,ave-bins))
        fac = (Nbins -1) / float(Nbins)
        err = np.sqrt(fac*sqsum)
    else:
        ave = np.mean(bins, axis=Jax)

        err = np.zeros(tL)
        for i in range(tL):
            sqsum = sum(map(lambda x:x*x,ave[i]-bins[:,i].real))
            fac = (Nbins -1) / float(Nbins)
            err[i] = np.sqrt(fac*sqsum)        
            
    return (ave,err)
#-------------------------------------------------------

def jackknife_binning(corr, binsize=1):
    
    Jax=0
    
    # Define the number of bins
    mod   = Ncfg%binsize
    Nbins = int((Ncfg - mod) / binsize)
    bins = np.zeros((Nbins,Nt))

    csum = np.sum(corr,axis=Jax) # Sum w.r.t to the configurations
    for t in range(Nt):
        for m in np.arange(1,mod+1):
            csum[t] -= corr[Ncfg-m,t]  # Throw away data in case there is modulo
            
        for b in range(Nbins):
            bsum = 0
            for k in range(binsize):
                bsum += corr[b*binsize+k,t]
                
            bins[b][t] = (csum[t] - bsum) / float(Ncfg - binsize - mod) # Bin averages for each bin,t

    return bins, Nbins
#-------------------------------------------------------

print('\nDone\n')
```

## Fit functions and relevant

```python
# Constant fit
def constFit(x,E0):
    return E0
#------------------------


# Ansatze for eigenvalues


# Fixed Z-factor fit
def Di(x,E0,En):
    return np.exp(-En*x)-np.exp(-E0*x)

def Mf(E0,En):
    return sum( map(lambda t:t, (ydata-np.exp(-E0*xdata))*Di(xdata,E0,En)/yerr**2) )

def Qf(E0,En):
    return sum( map(lambda t:t*t, Di(xdata,E0,En)) )

def Zf(E0,En):
    Mdef = Mf(E0,En)
    Qdef = Qf(E0,En)
    Zdef = Mdef/Qdef
    print('Zf: E0, En, M, Q, Z = ', E0, En, Mdef, Qdef, Zdef)
    return Zdef

def lambdaFz(x,E0,En):
    return np.exp(-E0*x) + Zf(E0,En)*( np.exp(-En*x) - np.exp(-E0*x) )
#------------------------
    

# Two-state fit
def lambdaF(x,E0,En,Z):
    return np.exp(-E0*x) + Z*( np.exp(-En*x) - np.exp(-E0*x) )
#    return np.exp(-E0*x)*(1-Z) + Z*np.exp(-En*x)

# 1-state fit
def lambdaF1s(x,E0,Z):
    return Z*np.exp(-E0*x)
#------------------------


# Chi-square
def chiSq(x, y, err, E0, En, Z):
    return sum( map(lambda t:t*t, (y-lambdaF(x,E0,En,Z))/err) )

def chiSq1s(x, y, err, E0, Z):
    return sum( map(lambda t:t*t, (y-lambdaF1s(x,E0,Z))/err) )

def chiSqFz(x, y, err, E0, En):
    return sum( map(lambda t:t*t, (y-lambdaFz(x,E0,En))/err) )


# More traditionally
def chiSq2(x, y, err, E0, En, Z):
    chi2 = 0
    for i in range(len(x)):
        chi2 += ( (y[i] - lambdaF(x[i],E0,En,Z))/err[i] )**2    
    return chi2
#------------------------


print('\nDone\n')
```

## Read correlators, construct effective energy

```python
effEnergy_Allbins = {}
effEnergy_Allave  = {}
corr2pt_Allbins   = {}
corr2pt_Allave    = {}

corr2pt_rowAvg = {}

corr2pt_rawAllave = {}

effEnergy_bins = {}
effEnergy_ave  = {}
corr2pt_bins   = {}
corr2pt_ave    = {}

corr2pt_rawAve = {}

Nbins_dict = {}
for pL in momLabels:
    runTag = '%s_%s'%(phaseInfo,pL)
    momVal, momTag = momInfo(pL)
    
    Ncfg = Ncfg_Dict[runTag]
    Nt = Nt_Dict[runTag]
    nRow = nRow_Dict[runTag]
    tSrcList = tSrc_Dict[runTag]
    NtSrc = len(tSrcList)

    for isrc, src in enumerate(optrList[runTag]):
        for isnk, snk in enumerate(optrList[runTag]):
            ckey = (pL,isrc,isnk)
            
            corr2pt_rawAll = np.zeros((Ncfg,Nt,nRow,NtSrc),dtype=np.complex128)

            for itSrc,tSrc in enumerate(tSrcList):
                tSrc_Tag = 't0_%d'%(tSrc)
                corr_file_pre='corr_2pt.baryon.n64.%s.%s' % (phaseInfo, tSrc_Tag)
                
                for ir,irow in enumerate(range(1,nRow+1)):
                    ckeyA = (pL,isrc,isnk,tSrc,irow)
                    corrFile = '%s/%s/%s/%s.src_%s_%d.snk_%s_%d.%s.dat' % (corr_dir,tSrc_Tag,momTag,corr_file_pre, src, irow, snk, irow, momTag)

                    with open(corrFile) as fp:
                        line = fp.readlines()
                        c = 0
                        for n in line:
                            it   = c%Nt
                            icfg = c//Nt
                            corr2pt_rawAll[icfg,it,ir,itSrc] = complex(np.float64(n.split()[1]),
                                                                       np.float64(n.split()[2]))
                            c += 1
                    # Reading file
                    
                    # Jackknife average on All two-point functions
                    corr2pt_Allbins[ckeyA], Nbins_dict[runTag] = jackknife_binning(corr2pt_rawAll[:,:,ir,itSrc].real)
                    Nbins = Nbins_dict[runTag]
                    corr2pt_Allave[ckeyA] = jackknife_ave(corr2pt_Allbins[ckeyA],Nbins,tL=Nt)

                    # Effective energy of All tSrc and rows
                    effEnergy_Allbins[ckeyA] = np.zeros((Nbins,Nt))
                    for it in range(Nt):
                        effEnergy_Allbins[ckeyA][:,it] = np.log(corr2pt_Allbins[ckeyA][:,it]/
                                                                corr2pt_Allbins[ckeyA][:,(it+1)%Nt])

                    effEnergy_Allave[ckeyA] = jackknife_ave(effEnergy_Allbins[ckeyA],Nbins,Nt)
                # for-irow

                # Standard average and standard error (still average over rows)
                ckeyB = (pL,isrc,isnk,tSrc)
                corr2pt_rowAvg[ckeyB] = np.mean(corr2pt_rawAll[:,:,:,itSrc],axis=2,dtype=np.complex128) # Average over rows
                corr2pt_rawAllave[ckeyB] = (np.mean(corr2pt_rowAvg[ckeyB],axis=0),
                                            np.std(corr2pt_rowAvg[ckeyB],axis=0)/np.sqrt(Ncfg))
                
            # for-tSrc
            
            # Average over rows and tSrc
            corr2pt_Tmp1 = np.mean(corr2pt_rawAll,axis=3,dtype=np.complex128) # over tSrc
            corr2pt_Tmp2 = np.mean(corr2pt_Tmp1  ,axis=2,dtype=np.complex128) # over rows
            
            # Standard average and standard error
            corr2pt_rawAve[ckey] = (np.mean(corr2pt_Tmp2,axis=0),np.std(corr2pt_Tmp2,axis=0)/np.sqrt(Ncfg))
            
            # Jackknife-binning and Effective energy on tSrc- and row-average            
            corr2pt_bins[ckey], Nbins = jackknife_binning(corr2pt_Tmp2.real)
            corr2pt_ave[ckey] = jackknife_ave(corr2pt_bins[ckey],Nbins,tL=Nt)
            
            effEnergy_bins[ckey] = np.zeros((Nbins,Nt))
            for it in range(Nt):
                effEnergy_bins[ckey][:,it] = np.log(corr2pt_bins[ckey][:,it]/
                                                    corr2pt_bins[ckey][:,(it+1)%Nt])

            effEnergy_ave[ckey] = jackknife_ave(effEnergy_bins[ckey],Nbins,tL=Nt)
    # for isrc/isnk
            
    print('%s done'%(pL))
            
print('\nDone\n')
```

### Construct Covariance Matrix of the correlator

```python

Ctld = {}
CovMat = {}

for pL in momLabels:
    runTag = '%s_%s'%(phaseInfo,pL)
    momVal, momTag = momInfo(pL)
    
    Ncfg = Ncfg_Dict[runTag]
    Nt = Nt_Dict[runTag]
    nRow = nRow_Dict[runTag]
    tSrcList = tSrc_Dict[runTag]
    NtSrc = len(tSrcList)

    for isrc, src in enumerate(optrList[runTag]):
        for isnk, snk in enumerate(optrList[runTag]):
            for itSrc,tSrc in enumerate(tSrcList):
                ckeyB = (pL,isrc,isnk,tSrc)

                C2spl = corr2pt_rowAvg[ckeyB] # (Ncfg x Nt)
                C2ave = corr2pt_rawAllave[ckeyB] # (2,Nt) -> (0,Nt): value, (1,Nt): error
                
                Ctld[ckeyB] = np.zeros((Ncfg,Nt))
                for ic in range(Ncfg):
                    for ti in range(Nt):
                        Ctld[ckeyB][ic,ti] = C2spl[ic,ti].real-C2ave[0][ti].real # sum(C - C_ave) for each t
    
            for ti in range(Nt):
                ckeyM = (pL,isrc,isnk,ti)
                CovMat[ckeyM] = np.zeros((NtSrc,NtSrc))
                
                for itSrcr,tSrcr in enumerate(tSrcList):
                    ckeyR = (pL,isrc,isnk,tSrcr)
                    rNrm = np.sqrt(Ctld[ckeyR][:,ti].dot(Ctld[ckeyR][:,ti]))
                    for itSrcc,tSrcc in enumerate(tSrcList):
                        ckeyC = (pL,isrc,isnk,tSrcc)
                        cNrm = np.sqrt(Ctld[ckeyC][:,ti].dot(Ctld[ckeyC][:,ti]))

                        CovMat[ckeyM][itSrcr,itSrcc] = (np.dot(Ctld[ckeyR][:,ti],Ctld[ckeyC][:,ti])/
                                                                    (rNrm*cNrm))

    print('%s done'%(pL))

print('\nDone\n')
```

## Perform simple constant fit on effective energy

```python

tini = 8
tfin = 13
tFit = np.arange(tini,tfin+1)
Ndata = len(tFit)

E0_fit = {}
E0_cov = {}

E0_fit_ave = {}
for pL in momLabels:
    momVal, momTag = momInfo(pL)
    runTag = '%s_%s'%(phaseInfo,pL)
    Nbins = Nbins_dict[runTag]
    nRow = nRow_Dict[runTag]
    
    for isrc, src in enumerate(optrList[runTag]):
        for isnk, snk in enumerate(optrList[runTag]):
            ckey = (pL,isrc,isnk)

            E0_fit[ckey] = np.zeros(Nbins)
            E0_cov[ckey] = np.zeros((Nbins,Ndata))

            E0_fitAll = np.zeros(Nbins)
            E0_covAll = np.zeros((Nbins,Ndata))

            nFit = 0
            for ib in range(Nbins):
                ydata = effEnergy_bins[ckey][ib,tini:tfin+1]
                yerr  = effEnergy_ave[ckey][1][tini:tfin+1] 
                if (not np.isnan(ydata).any()):
                    E0_fitAll[nFit], E0_covAll[nFit] = scipyOpt.curve_fit(constFit,tFit,ydata,sigma=yerr)
                    nFit += 1

            E0_fit[ckey] =  E0_fitAll[:nFit]

            if (np.shape(E0_fit[ckey])==(0,)) or (np.shape(E0_fit[ckey])==(1,)):
                E0_fit_ave[ckey] = (0,0)
            else:
                E0_fit_ave[ckey] = jackknife_ave(E0_fit[ckey],Nbins=nFit,tL=1)

    print('%s Done'%(pL))
    
print('\nDone\n')

```

## Perform GEVP

Perform GEVP for various t0

Solve the system $C(p;t)v(p;t,t0) = \lambda(p;t,t0) C(p;t0)v(p;t,t0)$ for each bin

Define $A = C(p,t)$, $B = C(p,t0)$

```python
corr2pt_Mat  = {}
corr2pt_GEVP = {}

corr2pt_rawAve[ckey]

# Construct symmetrized correlator for GEVP
for pL in momLabels:
    runTag = '%s_%s'%(phaseInfo,pL)
    Nt = Nt_Dict[runTag]
    Nbins = Nbins_dict[runTag]
    Nop = Nop_dict[runTag]
    
    corr2pt_Mat[pL] = np.zeros((Nbins,Nt,Nop,Nop))
    for isrc, src in enumerate(optrList[runTag]):
        for isnk, snk in enumerate(optrList[runTag]):
            ckey = (pL,isrc,isnk)
            corr2pt_Mat[pL][:,:,isrc,isnk] = corr2pt_bins[ckey]
    # src, snk

    # Symmetrize the correlator matrix for each bin and time
    corr2pt_GEVP[pL] = np.zeros((Nbins,Nt,Nop,Nop))
    for ib in range(Nbins):
        for t in range(Nt):
            corr2pt_GEVP[pL][ib,t] = 0.5 * (corr2pt_Mat[pL][ib,t] + 
                                            corr2pt_Mat[pL][ib,t].conj().T)
#-------------------------------------------

t0_range = [2,3,4,5]
tRngD = {2:(1,14),
         3:(1,14),
         4:(2,14),
         5:(2,14)}

eVals = {}
eVecs = {}
eVals_srtV = {}
eVals_srtV_ave = {}
Eeff_bins = {}
Eeff_ave  = {}
tRange ={}
for pL in momLabels:
    runTag = '%s_%s'%(phaseInfo,pL)
    Nt = Nt_Dict[runTag]
    Nbins = Nbins_dict[runTag]
    Nop = Nop_dict[runTag]

    for t0 in t0_range:
        tiG = tRngD[t0][0]
        tfG = tRngD[t0][1]
        tRange[t0] = np.arange(tiG,tfG+1)
        Npt = len(tRange[t0])

        eVals[(pL,t0)] = np.zeros((Nbins,Npt,Nop))
        eVals_srtV[(pL,t0)] = np.zeros((Nbins,Npt,Nop))
        eVecs[(pL,t0)] = np.zeros((Nbins,Npt,Nop,Nop))
        for tc, t in enumerate(tRange[t0]):
            for ib in range(Nbins):
                A = corr2pt_GEVP[pL][ib,t]
                B = corr2pt_GEVP[pL][ib,t0]

                eVals[(pL,t0)][ib,tc], eVecs[(pL,t0)][ib,tc] = la.eigh(A,B)
                
                # Sort eigenvalues by value (at t==t0, eVals = 1)
                if t<t0:
                    eVals_srtV[(pL,t0)][ib,tc] = np.sort(eVals[(pL,t0)][ib,tc]) # ascending
                elif t>=t0:
                    eVals_srtV[(pL,t0)][ib,tc] = np.sort(eVals[(pL,t0)][ib,tc])[::-1] # descending
                    
        # Effective evergy
        Eeff_bins[(pL,t0)] = np.zeros((Nbins,Npt,Nop))
        for tc, t in enumerate(tRange[t0][:-1]):
            for ib in range(Nbins):
                Eeff_bins[(pL,t0)][ib,tc] = np.log(eVals_srtV[(pL,t0)][ib,tc]/
                                                   eVals_srtV[(pL,t0)][ib,tc+1])

        for iop in range(Nop):
            Eeff_ave[(pL,t0,iop)] = jackknife_ave(Eeff_bins[(pL,t0)][:,:,iop],Nbins,tL=Npt)
            eVals_srtV_ave[(pL,t0,iop)] = jackknife_ave(eVals_srtV[(pL,t0)][:,:,iop],Nbins,tL=Npt)
                        
            
#print(eVals[('pz.-4',3)][30,1] - eVals_srtV[('pz.-4',3)][30,1])
#print(eVals[('pz.-4',3)][30,3] - eVals_srtV[('pz.-4',3)][30,3])
#print(eVals_srtV[('pz.-4',3)])

print('tRange = ', tRange)

print ('\nDone\n')            
```

## Perform fit of the eigenvalues

#### Including $E_0$, $E_n$ and $Z$

```python

covMat  = {}
En = {}
E0 = {}
Z = {}

tFrng = {}

E0_ave = {}
En_ave = {}
Z_ave  = {}

t0Try = 4
nOpTry = [0,1,2,3,4]

nFitG = {}

for pL in momLabels: #['pz.-4']:
    runTag = '%s_%s'%(phaseInfo,pL)
    Nt = Nt_Dict[runTag]
    Nbins = Nbins_dict[runTag]
    Nop = Nop_dict[runTag]

    for t0 in t0_range: #[t0Try]:
        tLmin = t0+4
        tLmax = tRngD[t0][1]
        
        tLminIdx = tLmin-tRngD[t0][0]
        tLmaxIdx = tLmax-tRngD[t0][0]+1
        
        tFrng[(pL,t0)] = np.arange(tLminIdx,tLmaxIdx)
        Ndata = len(tFrng[(pL,t0)])
        for iop in nOpTry: #range(Nop): #[0,1,2]:#
            ekey = (pL,t0,iop)
            
            En_all  = np.full(Nbins,None)
            E0_all  = np.full(Nbins,None)
            Z_all   = np.full(Nbins,None)
            chi_all = np.full(Nbins,None)

            iFitG = 0
            for ib in range(Nbins): #range(5):
                xdata = tFrng[(pL,t0)]
                ydata = eVals_srtV[(pL,t0)][ib,tLminIdx:tLmaxIdx,iop]
                yerr  = eVals_srtV_ave[ekey][1][tLminIdx:tLmaxIdx]
                try:
                    fitVals, covMat[ekey] = scipyOpt.curve_fit(lambdaF,
                                                               xdata, ydata,
                                                               sigma=yerr,
                                                               absolute_sigma=True)
                    En_all[iFitG] = fitVals[0]
                    E0_all[iFitG] = fitVals[1]
                    Z_all[iFitG]  = fitVals[2]

                    chi_all[iFitG] = chiSq(tFrng[(pL,t0)], ydata, yerr,
                                           En_all[iFitG], E0_all[iFitG], Z_all[iFitG]) / (Ndata-3-1)

                    iFitG += 1
                except:
                    continue
            # for Nbins
            
            nFitG[ekey] = iFitG
            
            E0[ekey] = E0_all[:iFitG]
            En[ekey] = En_all[:iFitG]
            Z[ekey]  = Z_all[:iFitG]
    
            if (np.shape(E0[ekey])==(0,)): # No successfull fit for this ekey
                E0_ave[ekey] = (None,None)
                En_ave[ekey] = (None,None)
                Z_ave[ekey]  = (None,None)
            elif (np.shape(E0[ekey])==(1,)): # Only one successfull fit
                E0_ave[ekey] = (E0[ekey],0)
                En_ave[ekey] = (En[ekey],0)
                Z_ave[ekey]  = (Z[ekey] ,0)
            else:
                E0_ave[ekey] = jackknife_ave(E0[ekey],Nbins=nFitG[ekey],tL=1)
                En_ave[ekey] = jackknife_ave(En[ekey],Nbins=nFitG[ekey],tL=1)
                Z_ave[ekey]  = jackknife_ave(Z[ekey] ,Nbins=nFitG[ekey],tL=1)
        # For-loop iop
        print('pL,t0 = ', pL, t0)

print('\nDone\n')

```

### Construct error bands

```python
NtBand = 201

TBandRng = {}

evalBand_val = {}
evalBand_err = {}
EnBand_val = {}
EnBand_err = {}

for pL in momLabels:
    runTag = '%s_%s'%(phaseInfo,pL)
    Nt = Nt_Dict[runTag]
    Nbins = Nbins_dict[runTag]
    Nop = Nop_dict[runTag]
    
    for t0 in t0_range:
        tBini = tFrng[(pL,t0)][0]
        tBfin = tFrng[(pL,t0)][-1]
        dtB = (tBfin-tBini)/(NtBand-1)
        TBandRng[(pL,t0)] = np.arange(tBini,tBfin+1e-8,dtB)
        
        for iop in nOpTry:
            ekey = (pL,t0,iop)
    
            evalBand_val[ekey] = np.zeros(NtBand)
            evalBand_err[ekey] = np.zeros(NtBand)
            EnBand_val[ekey]   = np.zeros(NtBand)
            EnBand_err[ekey]   = np.zeros(NtBand)
            
            if nFitG[ekey] == 0:
                evalBand_val[ekey] = np.full(NtBand,None)
                evalBand_err[ekey] = np.full(NtBand,None)
            elif nFitG[ekey] == 1:
                for itB,tBnd in enumerate(TBandRng[(pL,t0)]):
                    evalBand_val[ekey][itB] = lambdaF(tBnd, En_ave[ekey], E0_ave[ekey], Z_ave[ekey])
                    evalBand_err[ekey][itB] = 0.0
            else:
                for itB,tBnd in enumerate(TBandRng[(pL,t0)]):
                    vBand = np.zeros(nFitG[ekey])
                    for ifB in range(nFitG[ekey]):
                        vBand[ifB] = lambdaF(tBnd, En[ekey][ifB], E0[ekey][ifB], Z[ekey][ifB])
                    vBand_ave = jackknife_ave(vBand,nFitG[ekey],tL=1)                    
                    evalBand_val[ekey][itB], evalBand_err[ekey][itB] = vBand_ave[0], vBand_ave[1]
                # For itB

                
            # End if
        # For-loop iop    
        print('(pL,t0)', pL, t0)
            
print('\nDone\n')
```
